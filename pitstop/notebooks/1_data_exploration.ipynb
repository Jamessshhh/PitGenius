```python
import fastf1
import pandas as pd
import os

# Enable cache (don‚Äôt re-download ‚Äî uses local data)
fastf1.Cache.enable_cache('../data/cache')

# Define seasons and sessions
seasons = [2023, 2024]
session_type = 'R'  # R = Race

all_data = []

for year in seasons:
    schedule = fastf1.get_event_schedule(year)
    for _, event in schedule.iterrows():
        try:
            print(f"üèÅ Loading {event['EventName']} {year} ...")
            session = fastf1.get_session(year, event['EventName'], session_type)
            session.load(laps=True, telemetry=False)
            laps = session.laps
            laps['Year'] = year
            laps['Event'] = event['EventName']
            all_data.append(laps)
        except Exception as e:
            print(f"‚ö†Ô∏è Could not load {event['EventName']} {year}: {e}")

# Combine all races into one DataFrame
all_laps = pd.concat(all_data, ignore_index=True)

print(f"\n‚úÖ Loaded {len(all_laps)} laps from {len(all_data)} races across {len(seasons)} seasons.")
all_laps.head()
```

```python
# üí• Step 5 ‚Äì Stint & Tyre Degradation Analysis (All Drivers & Races)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Make sure all_laps is already loaded

# ‚úÖ Select only the needed columns
stint_data = all_laps[['Year', 'Event', 'Driver', 'Team', 'LapNumber', 'Compound', 'LapTime', 'Stint']].copy()
stint_data = stint_data.dropna(subset=['LapTime', 'Compound', 'Stint'])

# Convert LapTime to seconds
stint_data['LapTime'] = stint_data['LapTime'].dt.total_seconds()

# ‚öôÔ∏è Compute average lap time per stint (across all races)
avg_stint_perf = (
    stint_data.groupby(['Year', 'Compound', 'Stint'])
    .agg({'LapTime': 'mean'})
    .reset_index()
)

# ‚ú® Plot overall tyre degradation trend (across all drivers/races)
sns.set_style("whitegrid")
plt.figure(figsize=(10,6))
sns.lineplot(data=avg_stint_perf, x='Stint', y='LapTime', hue='Compound', style='Year', markers=True)
plt.title("üí• Average Lap Time vs Stint Number (All Races & Drivers)")
plt.xlabel("Stint Number")
plt.ylabel("Average Lap Time (seconds)")
plt.legend(title="Tyre Compound & Year")
plt.tight_layout()
plt.show()

# üèéÔ∏è Compare degradation for each team
team_stint_perf = (
    stint_data.groupby(['Year', 'Team', 'Compound', 'Stint'])
    .agg({'LapTime': 'mean'})
    .reset_index()
)

plt.figure(figsize=(12,6))
sns.lineplot(data=team_stint_perf, x='Stint', y='LapTime', hue='Team', style='Compound')
plt.title("üèéÔ∏è Average Lap Time per Team (Tyre Degradation Across Stints)")
plt.xlabel("Stint Number")
plt.ylabel("Average Lap Time (seconds)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Team / Tyre')
plt.tight_layout()
plt.show()

# üìä Summary: Average lap time per compound and year
compound_summary = (
    avg_stint_perf.groupby(['Year', 'Compound'])
    .agg({'LapTime': ['mean', 'std', 'count']})
    .round(2)
)

print("üìä Tyre Compound Performance Summary Across All Drivers & Races:")
display(compound_summary)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# --- Prepare the data ---
deg_data = all_laps[['Year', 'Event', 'Driver', 'Team', 'LapNumber', 'Compound', 'LapTime', 'Stint']].copy()
deg_data = deg_data.dropna(subset=['LapTime', 'Compound', 'Stint'])
deg_data['LapTimeSeconds'] = deg_data['LapTime'].dt.total_seconds()

# --- Compute average lap time per stint and compound ---
avg_laps = (
    deg_data.groupby(['Year', 'Compound', 'Stint'])
    .agg({'LapNumber': 'mean', 'LapTimeSeconds': 'mean'})
    .reset_index()
)

# --- Fit a linear regression to estimate degradation per compound ---
results = []
for year in avg_laps['Year'].unique():
    for compound in avg_laps['Compound'].unique():
        subset = avg_laps[(avg_laps['Year'] == year) & (avg_laps['Compound'] == compound)]
        if len(subset) > 2:
            X = subset[['LapNumber']]
            y = subset['LapTimeSeconds']
            model = LinearRegression().fit(X, y)
            slope = model.coef_[0]
            results.append({
                'Year': year,
                'Compound': compound,
                'DegradationRate (s/lap)': round(slope, 4)
            })

deg_rate_df = pd.DataFrame(results)
print("üßÆ Estimated Degradation Rate (seconds per lap):")
print(deg_rate_df.sort_values(['Year', 'Compound']))

# --- Visualize degradation rate per tyre ---
plt.figure(figsize=(8,5))
sns.barplot(data=deg_rate_df, x='Compound', y='DegradationRate (s/lap)', hue='Year', palette='coolwarm')
plt.title("üìâ Tyre Degradation Rate by Compound (2023‚Äì2024)")
plt.ylabel("Degradation per Lap (seconds)")
plt.xlabel("Tyre Compound")
plt.tight_layout()
plt.show()
```

```python
# üèÅ Step 7 ‚Äì Pit Stop Strategy Efficiency Analysis (All Races & Drivers)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
sns.set_palette("Set2")

# --- 1Ô∏è‚É£ Prepare pit stop data ---
pit_data = all_laps[['Year', 'Event', 'Driver', 'Team', 'PitInTime', 'PitOutTime', 'Compound', 'LapNumber']].copy()
pit_data = pit_data.dropna(subset=['PitInTime', 'PitOutTime'])

# Calculate pit stop duration in seconds
pit_data['PitDuration'] = (pit_data['PitOutTime'] - pit_data['PitInTime']).dt.total_seconds()

# --- 2Ô∏è‚É£ Average Pit Stops per Driver per Race ---
pit_counts = (
    pit_data.groupby(['Year', 'Event', 'Driver', 'Team'])
    .size()
    .reset_index(name='PitStops')
)

avg_pit_per_driver = (
    pit_counts.groupby(['Year', 'Driver'])
    ['PitStops']
    .mean()
    .reset_index()
)

plt.figure(figsize=(14,6))
sns.barplot(data=avg_pit_per_driver, x='Driver', y='PitStops', hue='Year', palette='viridis')
plt.title("üèéÔ∏è Average Pit Stops per Driver (All Races Combined)")
plt.xlabel("Driver")
plt.ylabel("Average Pit Stops")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# --- 3Ô∏è‚É£ Average Pit Stops per Team per Year ---
avg_pit_per_team = (
    pit_counts.groupby(['Year', 'Team'])
    ['PitStops']
    .mean()
    .reset_index()
)

plt.figure(figsize=(12,6))
sns.barplot(data=avg_pit_per_team, x='Team', y='PitStops', hue='Year', palette='coolwarm')
plt.title("üî© Average Pit Stops per Team per Year")
plt.xlabel("Team")
plt.ylabel("Average Pit Stops")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# --- 4Ô∏è‚É£ Average Pit Duration per Team ---
avg_duration = (
    pit_data.groupby(['Year', 'Team'])
    ['PitDuration']
    .mean()
    .reset_index()
)

plt.figure(figsize=(12,6))
sns.barplot(data=avg_duration, x='Team', y='PitDuration', hue='Year', palette='mako')
plt.title("‚è±Ô∏è Average Pit Stop Duration per Team (2023‚Äì2024)")
plt.xlabel("Team")
plt.ylabel("Average Duration (seconds)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("‚úÖ Step 7 Complete: Generated pit stop efficiency metrics for all drivers and teams across both seasons.")
```

```python
#Step 8 ‚Äì Race Pace Consistency & Performance Index
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# --- Prepare clean lap data ---
pace_data = all_laps[['Year', 'Event', 'Driver', 'Team', 'LapNumber', 'LapTime', 'Compound']].copy()
pace_data = pace_data.dropna(subset=['LapTime'])
pace_data['LapTimeSeconds'] = pace_data['LapTime'].dt.total_seconds()

# Remove unrealistic lap times (like in/out laps or red flag laps)
pace_data = pace_data[(pace_data['LapTimeSeconds'] > 60) & (pace_data['LapTimeSeconds'] < 120)]

# --- Calculate consistency metrics per driver ---
driver_stats = (
    pace_data.groupby(['Year', 'Driver'])
    .agg(
        AvgLapTime=('LapTimeSeconds', 'mean'),
        StdLapTime=('LapTimeSeconds', 'std'),
        TotalLaps=('LapNumber', 'count')
    )
    .reset_index()
)

# Calculate consistency score (lower std = better consistency)
driver_stats['ConsistencyIndex'] = 1 / (1 + driver_stats['StdLapTime'])

print("üèéÔ∏è Top 10 Most Consistent Drivers:")
print(driver_stats.sort_values('ConsistencyIndex', ascending=False).head(10))

# --- Visualization 1: Race pace consistency ---
plt.figure(figsize=(14,6))
sns.barplot(
    data=driver_stats.sort_values('ConsistencyIndex', ascending=False),
    x='Driver', y='ConsistencyIndex', hue='Year', palette='viridis'
)
plt.title("üèÅ Race Pace Consistency Index (2023 vs 2024)")
plt.xlabel("Driver")
plt.ylabel("Consistency Score (Higher = More Consistent)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# --- Visualization 2: Average Lap Time Comparison ---
plt.figure(figsize=(14,6))
sns.barplot(
    data=driver_stats.sort_values('AvgLapTime'),
    x='Driver', y='AvgLapTime', hue='Year', palette='coolwarm'
)
plt.title("‚ö° Average Race Lap Times per Driver (2023 vs 2024)")
plt.xlabel("Driver")
plt.ylabel("Average Lap Time (s)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Merge tyre degradation and pit data if you have them, otherwise skip those columns
# For now we'll focus on speed + consistency

# --- Normalize metrics ---
driver_stats['NormAvgLapTime'] = (driver_stats['AvgLapTime'] - driver_stats['AvgLapTime'].min()) / (driver_stats['AvgLapTime'].max() - driver_stats['AvgLapTime'].min())
driver_stats['NormConsistency'] = (driver_stats['ConsistencyIndex'] - driver_stats['ConsistencyIndex'].min()) / (driver_stats['ConsistencyIndex'].max() - driver_stats['ConsistencyIndex'].min())

# --- Compute OPI (you can tweak the weights) ---
driver_stats['OPI'] = (0.6 * (1 - driver_stats['NormAvgLapTime'])) + (0.4 * driver_stats['NormConsistency'])

# Sort and display
opi_sorted = driver_stats.sort_values('OPI', ascending=False)
print("üèÜ Overall Performance Index (Top 10 Drivers):")
print(opi_sorted[['Year', 'Driver', 'OPI']].head(10))

# --- Visualization ---
plt.figure(figsize=(14,6))
sns.barplot(data=opi_sorted, x='Driver', y='OPI', hue='Year', palette='mako')
plt.title("üèÅ Overall Driver Performance Index (2023 vs 2024)")
plt.xlabel("Driver")
plt.ylabel("Performance Index (Higher = Better)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# --- Step 10: Strategy Efficiency & Race Intelligence Analysis ---

# 1Ô∏è‚É£ Average lap time by tyre compound
compound_perf = (
    all_laps.groupby(['Year', 'Driver', 'Compound'])
    .agg({'LapTime': 'mean'})
    .reset_index()
)
compound_perf['LapTimeSeconds'] = compound_perf['LapTime'].dt.total_seconds()

# 2Ô∏è‚É£ Stint efficiency ‚Äî longer stints with good pace
stint_eff = (
    all_laps.groupby(['Year', 'Driver', 'Stint'])
    .agg({'LapNumber': 'count', 'LapTime': 'mean'})
    .reset_index()
)
stint_eff.rename(columns={'LapNumber': 'StintLength', 'LapTime': 'AvgLapTime'}, inplace=True)
stint_eff['AvgLapTimeSec'] = stint_eff['AvgLapTime'].dt.total_seconds()

# Normalize both metrics
stint_eff['NormPace'] = 1 - (stint_eff['AvgLapTimeSec'] - stint_eff['AvgLapTimeSec'].min()) / (
    stint_eff['AvgLapTimeSec'].max() - stint_eff['AvgLapTimeSec'].min()
)
stint_eff['NormStintLength'] = (stint_eff['StintLength'] - stint_eff['StintLength'].min()) / (
    stint_eff['StintLength'].max() - stint_eff['StintLength'].min()
)

# 3Ô∏è‚É£ Strategy Efficiency Score (SES)
stint_eff['SES'] = (0.6 * stint_eff['NormPace']) + (0.4 * stint_eff['NormStintLength'])

# Aggregate per driver
driver_ses = stint_eff.groupby(['Year', 'Driver'])['SES'].mean().reset_index()

# Merge with Overall Performance Index (from previous step)
race_intelligence = pd.merge(opi_sorted, driver_ses, on=['Year', 'Driver'], how='left')
race_intelligence['RaceIntelligenceIndex'] = (0.7 * race_intelligence['OPI']) + (0.3 * race_intelligence['SES'])

# Sort and show top drivers
top_intel = race_intelligence.sort_values('RaceIntelligenceIndex', ascending=False)
print("üèÜ Race Intelligence Index ‚Äì Top 10 Drivers:")
print(top_intel[['Year', 'Driver', 'RaceIntelligenceIndex']].head(10))

# Visualization
plt.figure(figsize=(14,6))
sns.barplot(data=top_intel, x='Driver', y='RaceIntelligenceIndex', hue='Year', palette='viridis')
plt.title("üß† Race Intelligence Index (Driver Skill + Strategy Efficiency)")
plt.xlabel("Driver")
plt.ylabel("Race Intelligence Index")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
